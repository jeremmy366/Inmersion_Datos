##Desafío 1

$1°$ Promedio del área de todos los inmuebles en los barrios en el dataset. El top 10

```python
import matplotlib.pyplot as plt

# Agrupar por barrio y calcular el área promedio
area_promedio_por_barrio = inmuebles.groupby('Barrio')['Area'].mean()

# Ordenar los barrios por precio promedio y seleccionar los 10 mejores
barrios_valoresAltos = area_promedio_por_barrio.sort_values(ascending=False).head(10)

# Calcular el promedio del área promedio para los 10 mejores barrios
promedio_areas = barrios_valoresAltos.mean()

# Imprimir el resultado
print('El promedio del área de los inmuebles en los 10 mejores barrios es:', promedio_areas)
print('El promedio de los valores altos de los inmuebles en los 10 mejores barrios es:\n', barrios_valoresAltos)
barrios_valoresAltos.plot(kind='bar', xlabel='Barrio', ylabel='Área Promedio', title='Top 10 Barrios por Área Promedio')

plt.show()
```

$$Conclusiones$$
1° El promedio del área de los inmuebles en los 10 mejores barrios es de 762.01 metros cuadrados. Esto sugiere que en estos barrios específicos, la oferta de inmuebles tiende a ser significativamente más grande en comparación con otros barrios en el conjunto de datos. Esto puede ser un indicio de que estos barrios son ideales para personas que buscan propiedades más espaciosas. $$$$

2° El barrio con el área promedio más alta es Modelia, con un impresionante promedio de 4316.75 metros cuadrados. Esta cifra destaca a Modelia como una zona con una oferta inmobiliaria excepcionalmente grande en comparación con los demás barrios del conjunto de datos. Si estás interesado en propiedades espaciosas, Modelia podría ser una opción a considerar en tu búsqueda de inmuebles.

$2°$ Consultar otros datos estadísticos, conteo, mediana, valores mínimo y máximo

```python
import matplotlib.pyplot as plt

# Cálculo de los estadísticos
conteo = inmuebles['Area'].count()
mediana = inmuebles['Area'].median()
minimo = inmuebles['Area'].min()
maximo = inmuebles['Area'].max()

# Imprimir los resultados
print('Conteo de observaciones:\n', conteo)
print('Mediana del área:\n', mediana)
print('Valor mínimo del área:\n', minimo)
print('Valor máximo del área:\n', maximo)

# Calcular y mostrar el gráfico de los 10 mejores promedios de área por barrio
promedio_area = inmuebles.groupby('Barrio')['Area'].median()
promedio_area.sort_values(ascending=False).head(10).plot.bar(color="blue")

plt.show()
```

$$Conclusiones$$
Basándonos en los resultados obtenidos, podemos concluir que el dataset contiene información sobre un total de 9,520 inmuebles, con áreas que abarcan desde 2 hasta 166,243 metros cuadrados. La mediana del área es de 80 metros cuadrados, lo que indica que la mitad de los inmuebles tienen un área menor o igual a 80 metros cuadrados, mientras que la otra mitad posee un área mayor. El valor mínimo del área es de 2 metros cuadrados, lo que sugiere la presencia de algunos inmuebles muy pequeños en el conjunto de datos.

Estos hallazgos son fundamentales para comprender la distribución de áreas de los inmuebles en el dataset. Además, proporcionan una referencia valiosa para evaluar la idoneidad de un inmueble en términos de su área en comparación con el conjunto de datos en su totalidad. Esto puede ser de gran utilidad al tomar decisiones relacionadas con la compra, venta o alquiler de inmuebles, permitiendo a los interesados considerar cómo se compara el tamaño de una propiedad en relación con otras disponibles en el mercado.

$2°$ Precio del m2 por barrio y hacer el gráfico más adecuado para esta nueva variable.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Conversión de tipos de datos
inmuebles['Area'] = inmuebles['Area'].astype(float)
inmuebles['Precio'] = inmuebles['Precio'].astype(float)

# Cálculo del precio por metro cuadrado
precio_m2 = inmuebles['Precio'] / inmuebles['Area']

# Número de barrios a mostrar
n_barrios = 10

# Cálculo del precio promedio por metro cuadrado por barrio y ordenamiento
precio_m2_por_barrio = precio_m2.groupby(inmuebles['Barrio']).mean().sort_values(ascending=False).head(n_barrios)

# Configuración de la paleta de colores personalizada
colores = sns.color_palette("Set2", len(precio_m2_por_barrio))

# Creación de la gráfica de barras utilizando Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x=precio_m2_por_barrio.index, y=precio_m2_por_barrio, palette=colores)

# Etiquetas y título
plt.xlabel('Barrio', fontsize=14)
plt.ylabel('Precio del metro cuadrado (millones de pesos colombianos)', fontsize=14)
plt.title('Precio del metro cuadrado por barrio en Bogotá', fontsize=16)
plt.xticks(rotation=45, fontsize=12)

# Agregar números en la parte superior de las barras
for i, v in enumerate(precio_m2_por_barrio):
    plt.text(i, v, f'{v:.2f}', ha='center', va='bottom', fontsize=10)

# Guardar la gráfica en formato png con un nombre descriptivo
plt.savefig('/content/drive/MyDrive/Alura_course2/valor_barrios.png', format='png')

# Ajuste automático del diseño para evitar solapamiento de elementos
plt.tight_layout()

plt.show()
```

$$Conclusiones$$
Hemos proporcionado información valiosa sobre la variación de los precios por metro cuadrado en los barrios de Bogotá. Puede ser útil para compradores, vendedores y profesionales del mercado inmobiliario que desean comprender mejor las tendencias de precios en la ciudad.
